## Remove NaN values

Remove rows that are mostly -1's (less than 100 non-Null values) and columns that have even one -1.

```python
def removenegatives(data=None, ipfilepath=None):

    if data is None:
        if  ipfilepath is not None:                
            with open(ipfilepath, 'rb') as fd:
                data = pkl.load(fd) 
        else:            
            return None
    
    # ROWS
    
    # This method works but leaves rows that are almost
    # all -1 except for isolated elements containing noise
    # Any not equal to -1 = Not all -1
    # i.e. remove rows with nothing but -1's
    # i.e. assign only rows that are not all -1's to q2
    # remove rows: axis = 1
    # q_pruned_rows = quotes[(quotes != -1).any(axis=1)]            
    # q_pruned_cols = q_pruned_rows.copy()
    
    # remove any row not containing a number of non-Null
    # values >= THRESHOLD, where THRESHOLD = 30% of the
    # number of columns. Only reason for using a percentage
    # is that occasionally a row will have a garbage value
    # the registers as non-NaN and a row that should be
    # deleted is left in the data and wreaks havoc.
    
    THRESHOLD = data.shape[1]
    maxNAcols = int(0.3 * THRESHOLD) 
    
    data = data.replace(-1, np.nan)
    q_pruned_rows = data.dropna(axis=0, thresh = maxNAcols) 
    q_pruned_cols = q_pruned_rows.copy()
    
    # COLUMNS
    
    # This loop works, but is not the best way to do it
    # for x in q_pruned_rows.columns:
    #     if ((q_pruned_rows[x] == -1).any()):
    #         q_pruned_cols = q_pruned_cols.drop(x, 1) 

    # q_pruned_cols = q_pruned_cols.replace(-1, np.nan)
    q_pruned_cols = q_pruned_cols.dropna(axis=1)            
    
    return q_pruned_cols
```

# Generate Returns

## Input = Clean Data

### Output = Returns

NOTE: `genreturns()` uses `removenegatives()` defined earlier in notebook.

```python
def genreturns(ipfolder, opfolder):
    
    picklefiles = list(glob(ipfolder + "*.pkl")) 

    for ipfilepath in picklefiles: 
        
        # remove folder from file path
        re_ipfolder = re.compile(ipfolder) 
        ipfile = re_ipfolder.sub('', ipfilepath) 
        
        # input and output files have the same name
        opfile = ipfile

        quotes = removenegatives(ipfilepath=ipfilepath)            
        
        # Calculate returns
        returns = quotes.pct_change().copy() 
        # Remove first row of NA values created by returns
        returns = returns.iloc[1:,:]
        
        opfilepath = opfolder + opfile
        with open(opfilepath, 'wb') as fd:
            pkl.dump(returns, fd)     
```

# Concatenate Data from Multiple Files

## Input = Dictionary of DFs with Close

### Output = Concatenated Close

```python
def concatclose(quotes):

    start = '1/1/2000'
    end = "1/1/2008"    

    dates = pd.date_range(start, end)
    df = pd.DataFrame(index=dates)

    l = list(quotes.keys())    
    
    for ticker in l:
        df1 = DataFrame(quotes[ticker]['Close'])
        df1 = df1.rename(columns={'Close': ticker})
        df = df.join(df1)

    return df 
```

### Output = MultiIndex Open-Close

```python
def concatopenclose(quotes):

    start = '1/1/2000'
    end = "1/1/2008"    

    # initialize dataframe with a stub column (to be removed later)
    dates = pd.date_range(start, end)
    df = pd.DataFrame(index=dates)  
    df['Open'] = -1
    df['Close'] = -1 
    df = DataFrame(df.stack()) 
    df = df.rename(columns={0: 'stubcol'})    

    l = list(quotes.keys())    
    
    for ticker in l:

        df1 = DataFrame(quotes[ticker][['Open', 'Close']])        
        df1 = DataFrame(df1.stack()).copy()       
        df1 = df1.rename(columns={0: ticker})
        df = df.join(df1)

    df = df.drop('stubcol', 1)
    
    return df 
```

### Glob Files and Call Concat Routines

```python
AGGREGATE_MARKET_CLOSE = 1
AGGREGATE_SECTORS_CLOSE = 2
AGGREGATE_SECTORS_OPENCLOSE = 3

def aggregatefiles(operation):
    
    if operation == AGGREGATE_MARKET_CLOSE:
        ipfolder = ip_market
        opfolder = closefolder_market
        con = concatclose        
    elif operation == AGGREGATE_SECTORS_CLOSE:
        # Aggregate close information 
        ipfolder = ip_sectors
        opfolder = closefolder_sectors
        con = concatclose
    elif operation == AGGREGATE_SECTORS_OPENCLOSE:
        # Aggregate open-close information 
        ipfolder = ip_sectors        
        opfolder = openclosefolder_sectors
        con = concatopenclose
    else:
        return 
    
    picklefiles = list(glob(ipfolder + "*.pkl")) 

    for ipfilepath in picklefiles: 

        # input dictionary of equity dataframes
        # and concatenate close price data

        with open(ipfilepath, 'rb') as fd:
            quotes = pkl.load(fd) 

        df = con(quotes)

        # pickle the output

        # remove folder from file path
        re_ipfolder = re.compile(ipfolder) 
        ipfile = re_ipfolder.sub('', ipfilepath) 

        # For entire market, leave descriptive suffix in file name
        if (operation != AGGREGATE_MARKET_CLOSE):
            # remove part of name not included in output
            # (if this suffix not present, then these two
            #  lines of code do nothing)
            re_output = re.compile(r'_2000-2008_len_2485')  
            ipfile = re_output.sub('', ipfile)   
        
        # File has same name in input and output folders
        opfile = ipfile

        opfilepath = opfolder + opfile

        with open(opfilepath, 'wb') as fd:
            pkl.dump(df, fd)         
```

## Apply Intrinio Filter to Returns Data

Sift Intrinio fundamental data to select data for tickers present in returns file.

`apply_intrinio_filter`:

Remove all stocks from returns data for which we do not have fundamental data.

```python
def apply_intrinio_filter(rtn_folder, rtn_file, fund_folder, fund_file):

    returns_filepath = rtn_folder + rtn_file
    fundamentals_filepath = fund_folder + fund_file
    
    with open(returns_filepath, 'rb') as fd:
        returns = pkl.load(fd)   

    # fundamentals = pd.read_csv(fundamentals_filepath, header=None)
    fundamentals = pd.read_excel(fundamentals_filepath)
    
    prices = fundamentals["Price to Earnings"].values
    tickers = fundamentals["Ticker"].values    
    
    tickers_fundamentals = set(tickers)
    tickers_prices = set(returns.columns)
    
    df_filtered_returns = pd.DataFrame()

    # step through return data testing each security
    # to see if its ticker is found in the fundamental
    # data provided by intrinio. Only save those stocks
    # for which we have fundamental data.
    
    for i, t in enumerate(tickers_prices):
        if t in tickers_fundamentals:
            df_filtered_returns[t] = returns[t]  
            
    opfilepath = rtn_folder + "returns_intrinio_filter_applied.pkl"
    
    with open(opfilepath, 'wb') as fd:
        pkl.dump(df_filtered_returns, fd)             
            
    return df
```

## Apply Price Filter to Fundamentals Data

```python
def apply_price_filter(rtn_folder, rtn_file, fund_folder, fund_file): 

    returns_filepath = rtn_folder + rtn_file
    fundamentals_filepath = fund_folder + fund_file
    
    with open(returns_filepath, 'rb') as fd:
        returns = pkl.load(fd)   

    fund = pd.read_excel(fundamentals_filepath)    
    
    tickers_prices = set(returns.columns)
    fundpruned = fund.loc[fund['Ticker'].isin(tickers_prices)] 
    
    opfilepath = rtn_folder + "fundamentals_price_filter_applied.pkl"
    
    with open(opfilepath, 'wb') as fd:
        pkl.dump(fundpruned, fd)     
    
    return fundpruned
```

```python
def retry(f, n_attempts=3):
    "Wrapper function to retry function calls in case of exceptions"
    def wrapper(*args, **kwargs):
        for i in range(n_attempts):
            try:
                return f(*args, **kwargs)
            except Exception:
                if i == n_attempts - 1:
                    raise
    return wrapper
```

```python
from pandas import date_range

start_date = datetime(2000, 1, 1).date()
end_date = datetime(2008, 1, 1).date()

def fetch_quandl_data(symbol, start_date, end_date):
    cols = ["Open", "High", "Low", "Close", "Volume"] 
    # NOTE how the params passed to quandl.get fall 
    # OUTSIDE the parentheses that enclose the arg
    # to retry!!!! The following gives an error:
    #      retry(quandl.get(symbol, 
    #            start_date=start_date, 
    #            end_date=end_date))
    dat = retry(quandl.get)(symbol, 
                start_date=start_date, 
                end_date=end_date)
    dat = dat[cols]    
    if len(dat) > 0:
        idx = date_range(start=start_date, end=end_date)
        # Index must be of type DatetimeIndex 
        # for the reindexing to work.
        dat.index = pd.DatetimeIndex(dat.index)
        dat = dat.reindex(idx, fill_value=-1)    
        # dat['Date'] = dat.index

    # return np.array(dat)
    return dat
```

## Fix WIKI Ticker Names

Remove `WIKI/` prefix from Quandl ticker symbols.

```python
re_match = re.compile('WIKI/')

def removeWIKIprefix(data=None, ipfilepath=None):

    if data is None:
        if  ipfilepath is not None:                
            with open(ipfilepath, 'rb') as fd:
                data = pkl.load(fd) 
                
        else:            
            return None     

    for hdr in data.columns:
        
        ticker = re_match.sub('', hdr)
        data = data.rename(columns={hdr: ticker})

    # if opfilepath != None:
    #     with open(opfilepath, "wb") as fd:
    #         pkl.dump(data, fd)          

    return data
```

## Get List of Tickers 

Save list of tickers included in the final returns file.

```python
def create_tickers_file(ipfilepath):
    
    with open(ipfilepath, 'rb') as fd:
        returns = pkl.load(fd)   

    tickers = list(returns.columns)
    
    # output file path
    re_pattern = re.compile(".pkl") 
    fn = re_pattern.sub("", ipfilepath)
    opfilepath = fn + "_tickers.pkl"    

    # save as pickle file
    with open(opfilepath, "wb") as fd:
        pkl.dump(tickers, fd)      
        
    # save as CSV file
    dftickers = pd.DataFrame(tickers)
    opfilepath = fn + "_tickers.csv"       
    dftickers.to_csv(opfilepath)
```
